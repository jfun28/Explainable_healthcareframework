```python
labeled_df=pd.read_excel("C:\\jupyter\\Elementary-student-Weight\\GAN_XAI\\data\\synthetic_data\\labeled_df.xlsx")
df_jeju=pd.read_csv("C:\jupyter\Elementary-student-Weight\GAN_XAI\data\제주하귀일초_dataset.csv",encoding="cp949")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming data is already loaded as df_seoul and df_jeju

# Features and labels
features = ['height', 'weight', 'step count', 'burned calorie', 'eat calorie', 'sleep time']
feature_labels = ['Height (cm)', 'Weight (kg)', 'Step Count', 'Burned Calories', 'Consumed Calories', 'Sleep Time (min)']

# Set visualization style
plt.figure(figsize=(15, 10))
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 15  # 전체 폰트 크기를 15로 설정

# Define custom color palette
custom_palette = {'A school (Seoul)': '#1f77b4', 'B school (Jeju)': '#ff7f0e'}  # Blue for Seoul, Orange for Jeju

# Plotting
plt.figure(figsize=(18, 12))
for i, (feature, label) in enumerate(zip(features, feature_labels)):
    plt.subplot(2, 3, i + 1)
    
    # Combine data
    seoul_df = pd.DataFrame({feature: labeled_df[feature], 'region': 'A school (Seoul)'})
    jeju_df = pd.DataFrame({feature: df_jeju[feature], 'region': 'B school (Jeju)'})
    combined_df = pd.concat([seoul_df, jeju_df])
    
    # Split violin plot with enhanced median visibility
    sns.violinplot(x='region', y=feature, data=combined_df, palette=custom_palette,
                  order=['A school (Seoul)', 'B school (Jeju)'], 
                  split=True, inner='box', saturation=0.7)
    
    # Add median values as text annotations with font size 15
    for idx, region in enumerate(['A school (Seoul)', 'B school (Jeju)']):
        median_val = combined_df[combined_df['region'] == region][feature].median()
        plt.text(idx, median_val, f'Median:\n{median_val:.1f}', 
                horizontalalignment='center', verticalalignment='bottom',
                fontsize=20)
    
    plt.title(f'{label} Comparison', fontsize=20)
    plt.ylabel(label, fontsize=20)
    plt.xlabel('Region', fontsize=20)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)

plt.tight_layout()
# plt.suptitle('Comparison of Data Features: A school (Seoul) vs B school (Jeju)', fontsize=15, y=1.02)
plt.show()
```

1. Height (cm)
중앙값: 제주가 서울보다 약간 높음.

분포: 두 지역 모두 비슷한 분포를 보이지만, 서울은 더 넓은 범위의 이상치를 가짐.

해석: 평균적인 신장 차이는 크지 않으며, 서울이 좀 더 다양한 키를 포함.

2. Weight (kg)
중앙값: 제주가 서울보다 높음.

분포: 서울은 이상치가 많고, 낮은 체중 분포가 더 많음.

해석: 제주의 체중은 전반적으로 더 높으며 서울은 저체중 비율이 많을 수 있음.

3. Step Count
중앙값: 제주가 서울보다 확실히 높음.

분포: 서울은 낮은 수치를 가진 분포가 많고 이상치가 더 많음.

해석: 제주의 일상 활동량(걸음 수)이 평균적으로 더 높음.

4. Burned Calories
중앙값: 서울이 더 높지만, 이상치가 매우 많음.

분포: 서울은 극단적으로 높은 칼로리 소모량이 많고, 제주에 비해 분산이 큼.

해석: 일부 매우 활발한 사용자가 서울에 집중되어 있거나 측정 오류 가능성 존재.

5. Consumed Calories
중앙값: 제주가 조금 더 높음.

분포: 두 지역 모두 이상치 존재. 서울이 더 넓은 분포.

해석: 식습관은 제주가 더 많이 섭취하는 경향이 있고 서울은 다양성이 큼.

6. Sleep Time (min)
중앙값: 제주가 서울보다 약간 높음.

분포: 제주가 더 고르게 분포되어 있고, 서울은 낮은 수면 시간이 많음.

해석: 제주가 평균적으로 더 많은 수면 시간을 확보.
